```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("../modified_Bank Customer Churn Prediction.csv", encoding ='utf-8')
```

```python
data.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
data.describe()
```

```python
data.describe(include='object')
```

```python
print(data['age'].mode())
print(data['age'].median())
print(data['age'].mean())
print('-'*10)
data['age'].describe()
```

```python
data['age'].value_counts()
```

```python
data['age'].value_counts().plot(kind='bar', figsize=(20, 10), title='Age Distribution')
```

```python
sns.boxplot(data['age'])
```

```python
for col in data.select_dtypes(include=["number"]).columns:
    print(f"Column: {col}")
    print(f"Minimum: {data[col].min()}")
    print(f"Maximum: {data[col].max()}")
    print(
        f"Mode: {data[col].mode()[0]}"
    )  # mode() returns a Series, take the first element
    print(f"Median: {data[col].median()}")
    print(f"Mean: {data[col].mean()}")
    print("-" * 10)
```
